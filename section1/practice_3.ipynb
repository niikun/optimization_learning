```python
import pulp
import numpy as np
```

```python
# 問題の定義   
prob = pulp.LpProblem("Maximize the profit", pulp.LpMaximize)

#　変数の定義
x = pulp.LpVariable("x",lowBound=0, cat="Integer")
y = pulp.LpVariable("y",lowBound=0, cat="Integer")
z = pulp.LpVariable("z",lowBound=0, cat="Integer")

# 目的関数の追加
prob += 5 * x + 3 * y + 2 * z

# 制約条件の追加
prob += 2 * x + z <= 5
prob += x + 2 * y <= 10
prob += y + z <= 8
prob += 2 * z <= 12

# 問題の解決
status = prob.solve()

# 結果の表示
print("Status:", pulp.LpStatus[status]) 
print("Result")
print("x:", pulp.value(x))
print("y:", pulp.value(y))
print("z:", pulp.value(z))
print("Objective value:", pulp.value(prob.objective))

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the limits for X, Y, and Z
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# Define the constraints
Z1 = (5 - 2 * X) / 0.1  # Constraint C1 (arbitrary scaling for visualization)
Z2 = (10 - X) / 2       # Constraint C2
Z3 = np.full_like(X, 8)  # Constraint C3
Z4 = np.full_like(X, 12)  # Constraint C4

# Set up the 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surfaces for each constraint
ax.plot_surface(X, Y, Z1, alpha=0.5, color='blue', label="C1")
ax.plot_surface(X, Y, Z2, alpha=0.5, color='red', label="C2")
ax.plot_surface(X, Y, Z3, alpha=0.5, color='green', label="C3")
ax.plot_surface(X, Y, Z4, alpha=0.5, color='yellow', label="C4")

# Label axes
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Constraints Visualization")

# Show the plot
plt.show()
```
